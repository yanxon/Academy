# Gaussian Process Regressor Note

Definition:
For any set S, a Gaussian Process on set S is a set of random variables ($Z_t: t \in S$) such that for any $n\in\mathbb{N}$, $\forall t_1, ..., t_n \in S$; $Z_{t_{1}}, ..., Z_{t_{n}}$ is multivariate Gaussian distribution. In another words, Gaussian process is a generalization of Gaussian distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kernel(a, b):
    dsq = np.sum(a**2, 1).reshape(-1, 1) + np.sum(b**2, 1) - 2.*np.dot(a, b.T)
    return np.exp(-0.5 * dsq)
```

```python
n = 100
X = np.linspace(-10, 10, n).reshape(-1, 1)
```

```python
K_ = kernel(X, X)
```

```python
L = np.linalg.cholesky(K_ + 1e-6*np.eye(n))
f_prior = np.dot(L, np.random.normal(size=(n,10)))
```

```python
plt.plot(X, f_prior)
```
