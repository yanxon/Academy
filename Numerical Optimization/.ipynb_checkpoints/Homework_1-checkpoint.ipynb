# Homework 1

Write the following functions and make contour plots.
1. Booth function
\begin{align}
f(x_1, x_2) = (x_1 + 2x_2 - 7)^2 + (2x_1 + x_2 - 5)^2
\end{align}
2. Branin function
\begin{align}
f(x_1, x_2) = a(x_2 - bx_1^2 + cx_1 - r)^2 + s(1-t)cos(x_1) + s\\
\textrm{where, } a = 1, b = 5.1/(4\pi^2), c = 5/\pi, r = 6, s = 10, t = 1/8\pi
\end{align}
3. Rosenbrock function
\begin{align}
f(x) = \sum^{n-1}_{i=1} \bigg[100(x_{i+1} - x_i^2)^2 + (1 - x_i)^2 \bigg]
\end{align}

```julia
using Gadfly
```

```julia
# Booth

booth(x1, x2) = (x1 + 2 * x2 - 7)^2 + (2 * x1 + x2 - 5)^2
arr = collect(-5:0.1:5)

println("The minimum Booth value when x1 = 1 and x2 = 3: ", booth(1, 3))

plot(z=(x,y)->booth(x, y), x=arr, y=arr, Geom.contour(levels=100))
```

```julia
# Branin

a = 1
b = 5.1/(4*pi^2)
c = 5/pi
r, s, t = 6, 10, 1/8/pi

branin(x1, x2) = a * (x2 - b*x1^2 + c*x1 - r)^2 + s * (1 - t) * cos(x1) + s
arr1 = collect(-5:0.1:10)
arr2 = collect(0:0.1:15)

plot(z=(x,y)->branin(x, y), x=arr1, y=arr2, Geom.contour(levels=50))
```

```julia
# Rosenbrock

function rosenbrock(x1, x2)
    return 100 * (x2 - x1^2)^2 + (1 - x1)^2
end

arr1 = collect(-1.5:0.1:2.0)
arr2 = collect(-0.5:0.1:3.0)

plot(z=(x,y)->rosenbrock(x, y), x=arr1, y=arr2, Geom.contour(levels=50))
```
