Sorting and Searching

```python
import time
import numpy as np
```

```python
#Bubble Sort

l = np.random.randint(1, 1000, 500)

t0 = time.time()
for i in range(len(l)):
    skip = 0
    for j in range(len(l)-1):
        if l[j+1] < l[j]:
            temp = l[j]
            l[j] = l[j+1]
            l[j+1] = temp
        else:
            skip += 1
t1 = time.time()

print(f"The time it takes to sort the list without skip is: {t1-t0}")

t2 = time.time()
for i in range(len(l)):
    skip = 0
    for j in range(len(l)-1):
        if l[j+1] < l[j]:
            temp = l[j]
            l[j] = l[j+1]
            l[j+1] = temp
        else:
            skip += 1
    if skip >= len(l)-1:
        break
t3 = time.time()

print(f"The time it takes to sort the list with skip is: {t3-t2}")

print(f"Sorting result: {l}")
```

```python
# Selection Sort

l = np.random.randint(1, 1000, 500)

for i in range(len(l)):
    temp = l[i]
    for j in range(i+1, len(l)):
        if l[j] < temp:
            temp = l[j]
            temp_idx = j
    if temp != l[i]:
        l[temp_idx] = l[i]
        l[i] = temp

print(l)
```

```python
# Merge Sort

l = list(np.random.randint(1, 1000, 500))

def merge(left, right):
    i, j = 0, 0
    result = []
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
            
    result += left[i:]
    result += right[j:]
    
    return result

def MergeSort(arr):
    if (len(arr) <= 1):
        return arr
    split = int(len(arr)/2)
    left = MergeSort(arr[:split])
    right = MergeSort(arr[split:])
    return merge(left, right)

newl = MergeSort(l)
print(newl)
```
